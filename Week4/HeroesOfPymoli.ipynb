##### Age Demographics
* The below each broken into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.) 
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
  * Normalized Totals


```python
# Create the bins in which Data will be held(Bins of 4 years i.e. <10, 10-14, 15-19, etc.))
bins = [0, 9, 14, 19, 24,29,34,39,50]

# Create the names for the bins
group_names = ['<10', '10-14', '15-19', '20-24','25-29','30-34','35-39', '40+']

#Convert ages to groups of age ranges
df['Bins'] = pd.cut(df["Age"], bins, labels=group_names,include_lowest=False)
```

```python
#Group the 'Purchase Count' by created 'Bins'
    ##(Returns Series)##
age_groups =df.groupby("Bins").count()['Item Name']

#Group the 'Average Purchase Price' by created 'Bins'
    ##(Returns Series)##
average_age_purchase_price=df.groupby(['Bins']).mean()['Price']

#Convert 'Purchase Count' series to dataframe
Age_Demographics = age_groups.to_frame()
#Rename column name from 'Item Name' to 'Purchase Count'
Age_Demographics.rename(columns={"Item Name": "Purchase count"},inplace = True)

#Convert 'Average Purchase Price' series to dataframe and add to existing dataframe
Age_Demographics['Average Purchase Price'] =average_age_purchase_price.to_frame()

#Calculate 'Total Purchase Value' and add to existing dataframe
Age_Demographics['Total Purchase Value'] = Age_Demographics['Purchase count']*Age_Demographics['Average Purchase Price']
Age_Demographics
```

##### Top Spenders

* Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
  * SN
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value

```python
#Group the 'Total Purchase Value' by 'SN' i.e Player Name
    ##(Returns Series)##
player_purchase_value =df.groupby("SN").sum()['Price']

#Group the 'Average Purchase Price' by 'SN' i.e Player Name
    ##(Returns Series)##
player_average_purchase_price=df.groupby(['SN']).mean()['Price']

#Convert 'Total Purchase Value' series to dataframe
top_spenders=player_purchase_value.to_frame()
#Rename column name from 'Price' to 'Total Purchase Value'
top_spenders.rename(columns={"Price":"Total Purchase Value"}, inplace= True)
#Calculate the 5 highest spenders by 'Total Purchase Value' and add to dataframe
top_5_spenders=top_spenders.sort_values(by= 'Total Purchase Value',ascending=False).head()

#Convert 'Average Purchase Price' series to dataframe and add to existing dataframe
top_5_spenders["Average Purchase Price"] =player_average_purchase_price.to_frame()

#Calculate 'Purchase Count', Convert datatype to integer and add to existing dataframe
top_5_spenders["Purchase Count"] = (top_5_spenders["Total Purchase Value"]/top_5_spenders["Average Purchase Price"]).astype(dtype=int)


#Reorder Columns
top_5_spenders=top_5_spenders[['Purchase Count','Average Purchase Price','Total Purchase Value']]
top_5_spenders
```

##### Most Popular Items

* Identify the 5 most popular items by purchase count, then list (in a table):
  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value

```python
#Group the 'Purchase Count' by 'Item Name' and 'Item ID'
    ##(Returns Series)##
item_purchase_count =df.groupby(["Item Name","Item ID"]).count()['Price']

#Group the Average Purchase Price by 'Item Name' and 'Item ID'
    ##(Returns Series)##
item_average_purchase_price=df.groupby(['Item Name',"Item ID"]).mean()['Price']

#Convert 'Purchase Count' series to dataframe
most_popular_items=item_purchase_count.to_frame()
#Rename column name from 'Price' to 'Purchase Count'
most_popular_items.rename(columns={"Price":"Purchase Count"}, inplace= True)
#Calculate the 5 highest spenders by Purchase Count and add to dataframe
top_5_popular_items=most_popular_items.sort_values(by= 'Purchase Count',ascending=False).head()

#Convert 'Average Purchase Price' series to dataframe and add to existing dataframe
top_5_popular_items["Average Purchase Price"] =item_average_purchase_price.to_frame()

#Calculate 'Total Purchase Value' and add to existing dataframe
top_5_popular_items["Total Purchase Value"] = top_5_popular_items["Purchase Count"]*top_5_popular_items["Average Purchase Price"]
top_5_popular_items 


```

##### Most Profitable Items

* Identify the 5 most profitable items by total purchase value, then list (in a table):
  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value

```python
#Group the 'Total Purchase Value' by 'Item Name' and 'Item ID'
    ##(Returns Series)##
item_purchase_value =df.groupby(["Item Name","Item ID"]).sum()['Price']

#Group the 'Average Purchase Price' by 'Item Name' and 'Item ID'
    ##(Returns Series)##
item_average_purchase_price=df.groupby(['Item Name',"Item ID"]).mean()['Price']

#Convert 'Total Purchase Value' series to dataframe
most_profitable_items=item_purchase_value.to_frame()
#Rename column name from 'Price' to 'Total Purchase Value'
most_profitable_items.rename(columns={"Price":"Total Purchase Value"}, inplace= True)
#Calculate the 5 most profitable items by 'Total Purchase Value' and add to dataframe
top_5_profitable_items=most_profitable_items.sort_values(by= 'Total Purchase Value',ascending=False).head()

#Convert 'Average Purchase Price' series to dataframe and add to existing dataframe
top_5_profitable_items["Average Purchase Price"] =item_average_purchase_price.to_frame()

#Calculate 'Purchase Count', Convert datatype to integer  and add to existing dataframe
top_5_profitable_items["Purchase Count"] = (top_5_profitable_items["Total Purchase Value"]/top_5_profitable_items["Average Purchase Price"]).astype(dtype=int)
top_5_profitable_items = top_5_profitable_items[['Purchase Count','Average Purchase Price','Total Purchase Value']]
top_5_profitable_items
```
